```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import math
```

```python
# 1. Setting Up Headers

header = {
    "User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}
```

```python
# 2. Making the Request


r = requests.get("https://www.last.fm/tag/top+tracks/tracks" , headers=header)
r.status_code
```

```python
# 3. Parsing the HTML Content

soup = BeautifulSoup(r.content, 'html.parser')
soup
```

```python
# 4. Extracting Data with a List Comprehension for one page 

songs_per_page = [(s.get_text(strip=True)) for s in soup.find_all("td", attrs={"class": "chartlist-name"})]
songs_per_page
```

```python
# Look at the number of songs per page

len(songs_per_page)
```

```python
# Extracting Data(song titles) with a List Comprehension for multiple pages (The Loop)

base_url = "https://www.last.fm/tag/top+tracks/tracks"

all_song_titles = []
num_pages = 7

for page in range(1 ,num_pages + 1):
    
    url = base_url + "?page" + str(page)
    r = requests.get(url , headers=header)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, 'html.parser')
        song_titles = [(s.get_text(strip=True)) for s in soup.find_all("td", attrs={"class": "chartlist-name"})]
        all_song_titles.extend(song_titles)
    
    else:
        print("Website not ok.")
    
    
    
all_song_titles  
    
```

```python
# Look at the total number of songs

len(all_song_titles)
```

```python
# Extracting Data(sartist names) with a List Comprehension for multiple pages (The Loop)

all_singer_names = []


num_pages = 7

for page in range(1 ,num_pages + 1):
    
    url = base_url + "?page" + str(page)
    r = requests.get(url , headers=header)
    
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, 'html.parser')
        singer_names = [(s.get_text(strip=True)) for s in soup.find_all("td", attrs={"class": "chartlist-artist"})]
        all_singer_names.extend(singer_names)
    
    else:
        print("Website not ok.")
        
    
all_singer_names  



```

```python
# Look at the toal number of singers

len(all_singer_names)
```

```python
# 5. Make a Dataframe with new lists


list(zip(all_singer_names , all_song_titles))
```

```python

Gnod_df = pd.DataFrame(zip(all_singer_names , all_song_titles), columns=["Artist Name", "Song Name"])

Gnod_df
```

```python
# 6.Prioritize the MVP

def song_recommender(song_and_artist):
    song_name,artist_name = song_and_artist.strip().lower().split(" - ")
    if ((Gnod_df['Song Name'].str.lower() == song_name) & (Gnod_df['Artist Name'].str.lower() == artist_name)).any():
        return "another song from Gnod "
    else:
        return "a song from somewhere else"
```

```python
song_and_artist = input('Write down a song and artist in the format "Song - Artist": ')


recommendation = song_recommender(song_and_artist)
recommendation
```
